```
# Import required file
import numpy as np
import matplotlib.pyplot as plt
import pandas as pb
import seaborn as sns





```

```
dataset  = pb.read_csv('Iris.csv')
Features_data = dataset.iloc[:,1:-1].values
Labled_output = dataset.iloc[:,-1:].values
Labled_output.shape

```

```
sns.pairplot(dataset.drop('Id', axis =1),hue = 'Species', height = 3)
```

```
from sklearn.model_selection  import  train_test_split


x_train, x_test, y_train, y_test = train_test_split(Features_data,Labled_output,random_state = 10, test_size= 0.2)

print(y_test.shape)
```

```
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()

y_train = encode.fit_transform( y_train)
y_test = encode.transform(y_test)
```

```
from sklearn.linear_model import LogisticRegression 
classifier1 = LogisticRegression()
classifier1.fit(x_train,y_train)
```

```
from sklearn.svm import SVC
classsifier2 = SVC()

classsifier2.fit(x_train,y_train)
```

```
from sklearn.ensemble import RandomForestClassifier
classsifier3 = RandomForestClassifier()

classsifier3.fit(x_train,y_train)
```

```
from sklearn.neighbors import KNeighborsClassifier
classsifier4 = KNeighborsClassifier()

classsifier4.fit(x_train,y_train)
```

```
from sklearn.metrics import accuracy_score
y_pred_logi = classifier1.predict(x_test)
logi_score = accuracy_score(y_pred_logi,y_test)

y_pred_svc = classsifier2.predict(x_test)
svc_score = accuracy_score(y_pred_svc,y_test)

y_pred_tree = classsifier3.predict(x_test)
tree_score = accuracy_score(y_pred_tree,y_test)

y_pred_knn =classsifier4.predict(x_test)
knn_score = accuracy_score(y_pred_knn,y_test)

```

```
accuracy_score = {"SCORE": [logi_score,svc_score,tree_score,knn_score]}
accuracy_data = pb.DataFrame(accuracy_score,index = ['Logistic ','Support vector Classifier','Random forest classifier','K near Neighbors'] )
accuracy_data.head()
```
